```python
import pandas as pd
import string
from english_words import get_english_words_set
```

```python
# user entries
main_letter = 'u'
sub_letters = ['altbme']
```

```python
words = get_english_words_set(['web2'], lower=True)
df = pd.DataFrame(words)
sub_letters = list(sub_letters[0])

print(sub_letters)
print(f"{len(sub_letters) = }")

drop_letters = list(string.ascii_lowercase)
drop_letters.remove(main_letter)
for sub in sub_letters:
    drop_letters.remove(sub)
    
print(drop_letters)
print(f"{len(drop_letters) = }")
```
